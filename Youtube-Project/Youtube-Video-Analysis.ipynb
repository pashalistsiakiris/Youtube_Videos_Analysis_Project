```python
#we see that it is taking too long time for kernel to respond because this is a very large dataset so we are going to take a sample of it
df2=df.head(2500).copy()
```

```python
lst=[]
for row in df2['comment_text']:
    try:
        lst.append(TextBlob(row).sentiment.polarity)
    except:
        lst.append(0)
```

```python
df2['polarity']=lst
```

```python
df2
```

```python
#now through polarity we will find which videos has the most positive sentiments
#we will consider a range from 0.7 to 1
df2[df2['polarity']>0.7].groupby('video_id').count()['polarity'].sort_values(ascending=False)
#we found the video with the most mositive sentiments
```

```python
#now we will find which words are used the most through wordcloud analysis for videos with polarity>0.7
from wordcloud import WordCloud,STOPWORDS
df3=df2[df2['polarity']>0.7]
```

```python
total_positive_comments=' '.join(list(df3['comment_text']))
```

```python
cloud=WordCloud(stopwords=set(STOPWORDS)).generate(total_positive_comments)
```

```python
plt.imshow(cloud)
plt.axis('off')
```

```python
#now lets do the same for negative sentiments
df_neg=df2[df2['polarity']<-0.7]
```

```python
total_negative_comments=' '.join(df_neg['comment_text'])
cloud2=WordCloud(stopwords=set(STOPWORDS)).generate(total_negative_comments)
```

```python
plt.imshow(cloud2)
plt.axis('off')
```

```python
#now what we are going to do is we will combine data from videos from various countries and perform analysis there 
#we will import those csvs and do a concatanation to produce a full dataframe
import os
files=os.listdir(r'C:\Users\pasxa\Desktop\github projects\Youtube-Project')
```

```python
files=[file for file in files if '.csv' in file]
```

```python
files.remove('UScomments.csv')
```

```python
files
```

```python
full_df=pd.DataFrame()
```

```python
for file in files:
    full_df=pd.concat([full_df,pd.read_csv(file,on_bad_lines='skip',encoding='iso-8859-1')])
```

```python
full_df
```

```python
#now having the full dataframe we are goin to find the most liked category
#to find that we must map the number in the category column into the category name 
#we will take it from the us.json file
json1=pd.read_json('US_category_id.json',encoding='iso-8859-1')
```

```python
json1['items'][0]
```

```python
dict1={}
```

```python
#we will create a dictionary with keys of 'id' and values of 'title'
for item in json1['items']:
    try:
        if item['id'] not in dict1.keys():
            dict1.update({int(item['id']):item['snippet']['title']})
    except:
        pass
```

```python
dict1
```

```python
full_df.info()
```

```python
full_df['category_id']=full_df['category_id'].apply(lambda x:dict1[x])
```

```python
full_df
```

```python
#now that we did the mapping we are going to see which is the most liked category
full_df.groupby('category_id')['likes'].sum().sort_values(ascending=False).head(5)
#we see that is the music which is reasonable
```

```python
#now with seaborn we will create some charts
```

```python
full_df.groupby('category_id')['likes'].sum().sort_values(ascending=False).head(5).plot.bar()
```

```python
plt.figure(figsize=(30,10))
sns.boxplot(x='category_id',y='likes',data=full_df,whis=1.5,flierprops=dict(marker='o',markersize=5,linestyle='none'))
plt.xlabel('category_id',fontsize=12)
#we have extremely high likes for some categories eg music
```

```python
sns.violinplot(x='category_id',y='likes',data=full_df[(full_df['category_id']=='Music') | (full_df['category_id']=='Entertainment')])
```

```python
#now lets analyze whether the audience is engaged or not
sns.scatterplot(x='views',y='likes',data=full_df)
#we see in this plit that as lond as the views are increasing the likes are increasing too
```

```python
sns.scatterplot(x='views',y='comment_count',data=full_df)
#we see that the same thing is happening with comments but with a much slower rate.People rarely will comment on a video than just push the 'like' button
```

```python
#now lets create a new feature called comments_per_likes and see which category has the highest
full_df['comments_per_like']=full_df['comment_count']/full_df['likes']
```

```python
plt.figure(figsize=(30,10))
sns.boxplot(x='category_id',y='comments_per_like',data=full_df)
plt.tight_layout()
```

```python
full_df['comments_per_view']=full_df['comment_count']/full_df['views']
```

```python
plt.figure(figsize=(30,10))
sns.boxplot(x='category_id',y='comments_per_view',data=full_df)
plt.tight_layout()
# we see that the most and highest outliers are in entairtnment videos ,which is reasonable
```

```python
#now lets see something else.Whether punctuation in title affects the likes or the views
#we will create a new column with the number of punctuation in their title
import string 
```

```python
full_df['number_of_punc']=full_df['title'].apply(lambda x:len([char for char in x if char in string.punctuation]))
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='number_of_punc',y='views',data=full_df)
#we see that we can achive the maximum number of views when the number of punc is between 1 andv 8.Higher number lead to immediate reduce of views.
```
